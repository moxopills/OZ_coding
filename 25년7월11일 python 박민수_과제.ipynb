__[문제3]<br>__

다음 리스트에서 최댓값과 최솟값, 그리고 평균을 출력해주세요.
<br>

```
nums = [1, 2, 3, 4, 5, 6, 7, 9, 10]
```



````python
num = [1,2,3,4,5,6,7,9,10]

max_num = max(num)
min_num = min(num)
average = sum(num)/len(num)

print("최댓값:", max_num)
print("최솟값:", min_num)
print("평균:", average)
````
